```python
label = dataset.iloc[:,[0, 1, 2]]  #[행인덱스, 열인덱스]
```

```python
X_train, X_test, y_train, y_test = train_test_split(data,
                                                    label,
                                                    test_size = 0.2,
                                                    shuffle = True,
                                                    random_state = 42)
```

```python
print('X_train: ', X_train.shape)
print('X_test: ', X_test.shape)
print('y_train: ', y_train.shape)
print('y_test: ', y_test.shape)
```

```python
#RF
from sklearn.ensemble import RandomForestClassifier
from sklearn.multioutput import MultiOutputClassifier
from sklearn import metrics
from sklearn.metrics import accuracy_score
```

```python
X_train1, X_val1, y_train1, y_val1 = train_test_split(X_train,
                                                    y_train,
                                                    test_size = 0.2,    #val로 뺄 거
                                                    shuffle = True,
                                                    random_state = 42)
```

```python
X_train1.shape
```

```python
X_val1.shape
```

```python
y_train1.shape
```

```python
y_val1.shape
```

```python
from sklearn.ensemble import RandomForestRegressor #RandomForestClassifier 오류

model_rf = RandomForestRegressor()
```

```python
model_rf = model_rf.fit(X_train1, y_train1)
```

```python
accuracy = model_rf.score(X_val1, y_val1)
print('RandomForest mean accuracy :', accuracy)
```

```python
features = X_train.columns
```

```python
importances = model_rf.feature_importances_
```

```python
import matplotlib.pyplot as plt
```

```python
indices = np.argsort(importances)[-9:]
plt.title('Feature importances')
plt.barh(range(len(indices)), importances[indices], color = 'b', align = 'center')
plt.yticks(range(len(indices)), [features[i]for i in indices])
plt.xlabel('Relative importance')
plt.show()
```
